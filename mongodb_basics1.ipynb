### Update in mongoDb

* .updateOne(filter,update) - if that particular field is not avilable in update statement then it will add that field into the that particular document

> $set - operator which is used to add and update the filed 

> $unset - operator which is used to remove the filed  

> db.hi.updateOne({name:'abhi'},{$set:{isgraduated:false}})

>  db.hi.updateOne({name:'abhi'},{$unset:{gpa:true}})

>  db.hi.updateMany({},{$set:{isgraduated:True}})

>  db.hi.updateMany({isgraduated:{$exists:false}},{$set:{isgraduated:true}})   - $exists - operation is used check if that specified field exist or not in a document  

### Delete one or many documents from the collection 

> db.hi.deleteOne({name:'abhi'}) - delete one document

> db.hi.deleteMany({isgraduated:true}) - delete many document 

> db.hi.deleteMany({age:{$exists:false}}) - delete the document which satisfied the condition exists

### Comparision Operators 

> denoted with $ 

> Comparision Operators return data based on the value comparision 



> db.hi.find({name:{$ne:'abhi'}})  = not equal to operator 

> dhirajdb> db.hi.find({name:{$eq:'abhi'}})  = equal to operator 

>  db.hi.find({age:{$gt:10}})  = greater than to operator 

>  db.hi.find({age:{$gte:10}}) = greater than or equal to operator 

>  db.hi.find({age:{$lt:10}}) = less than to operator 

>  db.hi.find({age:{$lte:10}}) = less than or equal to operator 

> db.hi.find({age:{$gte:12,$lte:29}}) = combine comparision operator to find the range of data 

> db.hi.find({age:{$in:[10,12,18,19]}}) = in operator,check if that field value present in the list

> dhirajdb> db.hi.find({age:{$nin:[10,12,18,19]}}) not in operator ,opposite of in operator 

### Logical Operator 

![Screenshot 2024-11-26 090010.png](<attachment:Screenshot 2024-11-26 090010.png>)

> logical operator returns data based on the evaluation of expressions to true or false 

> db.hi.find({$and:[{name:'abhi'},{age:{$lte:19}}]}) = and logical operator ,both expression need to true  

> db.hi.find({$or:[{name:'abhi'},{age:{$lte:19}}]})  = or logical operator ,atleast one expression need to true  

> db.hi.find({$nor:[{name:'abhi'},{age:{$lte:19}}]}) = nor logical operator,both expression need to false 

> db.hi.find({$or:[{name:'abhi'},{age:{$not:{$lte:19}}}]})  = not logical operator,makes the expression which is true to false and viceversa
